# Power Transmission Control with Stochastic Augmented Lagrangian

```python
import numpy as np
```

```python
class pwrTxControl:
    def __init__(self,initPwr,stepSize,muInit,c,errorLimit):
        self.pwr = initPwr
        self.stepSize = stepSize
        self.mu = muInit
        self.c = c
        self.errorLimit = errorLimit
    
    def BER(self):
        self.ber = (1/self.pwr)
        return self.ber
        
    def updatePwr(self):
        self.pwr = self.pwr-self.stepSize*(1+self.mu+self.c*self.BER()-self.errorLimit)
        print(self.pwr)
    
    def updateMu(self):
        self.mu = max([self.mu+self.stepSize*self.ber-self.errorLimit,0])
```

```python
pwrTxCtrlModel = pwrTxControl(1,0.01,0,1,0.01)
for i in range(100):
    pwrTxCtrlModel.updatePwr()
    pwrTxCtrlModel.updateMu()
```
